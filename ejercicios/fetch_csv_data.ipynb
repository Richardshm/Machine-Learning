```python
#librarys
import os
import tarfile
import urllib
import pandas as pd

%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
#Link to Github directory where the .csv file is located
PATH = 'https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/'
DOWNLOAD_ROOT = PATH + 'lifesat/oecd_bli_2015.csv'

#Local path to save the file
SAVE_PATH = 'datasets'
```

```python
#Function to fetch csv 
def fetch_csv_data(download_root = DOWNLOAD_ROOT, save_path = SAVE_PATH):
    #make dir if no exist
    if not os.path.isdir(save_path):
        os.makedirs(save_path)
    
    #Path where the .csv file will be saved
    file = save_path+'/oecd_bli_2015.csv'
    #Download the .csv from github to the local directory
    urllib.request.urlretrieve(download_root, file)

    
fetch_csv_data()
```

```python
#Function to read .csv in local path
def load_csv_data(save_path=SAVE_PATH):
    #Path where is saved the .csv
    file = save_path+'/oecd_bli_2015.csv'
    return pd.read_csv(file)

data = load_csv_data()
```
