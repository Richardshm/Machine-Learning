```python
#Librerias
import os
import tarfile
import urllib
import pandas as pd

%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
#Link de directorio de Github donde se encuentra el archivo .csv
PATH = 'https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/'
DOWNLOAD_ROOT = PATH + 'lifesat/oecd_bli_2015.csv'

#Path local donde guardar el archivo
SAVE_PATH = 'datasets'
```

```python
def fetch_csv_data(download_root = DOWNLOAD_ROOT, save_path = SAVE_PATH):
    #Crea la carpeta datasets en caso que no exista
    if not os.path.isdir(save_path):
        os.makedirs(save_path)
    
    #Path donde se guardara el archivo tgz
    file = save_path+'/oecd_bli_2015.csv'
    #Se descarga el tgz de github en el dirrectorio local
    urllib.request.urlretrieve(download_root, file)

    
fetch_csv_data()
```

```python
def load_csv_data(save_path=SAVE_PATH):
    #Path donde esta el archivo csv
    file = save_path+'/oecd_bli_2015.csv'
    return pd.read_csv(file)

data = load_csv_data()
```
